```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["font.family"] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
url = 'https://nba.hupu.com/standings'
nba_data = pd.read_html(url)  #header=0 去掉第一行列索引

nba_data
```

```python
df1 = nba_data[0].loc[0:16, :]
df1.columns = df1.iloc[1]
df1 = df1.iloc[2:, :]

df2 = nba_data[0].loc[17:, :]
df2.columns = df2.iloc[1]
df2 = df2.iloc[2:, :]

df1

df2
```

```python
# 将df1增加一列"所属联盟"，赋值为"东部赛区"
# 将df2增加一列"所属联盟"，赋值为"西部赛区"
# 将df1、df2合并成df

df1['所属联盟'] = '东部赛区'
df2['所属联盟'] = '西部赛区'

df = pd.concat([df1, df2])

df
```

```python
# 将df数据保存为csv文件 nba.csv

df.to_csv('nba.csv', encoding='utf-8')
```

```python
# 将所有球队按照 胜率(百分比) 降序排序，排名相同的队伍按 落后场次 升序排名
# 提示：ascending也可以接收列表参数
df = df.sort_values(['胜率', '净胜'], ascending=[False, False])
df
```

```python
# 对df按“所属联盟”分组，计算胜、负、胜率、净胜的最大值、最小值、均值、标准差
# 注意：胜率不是数值，要转换成小数

df['胜率新'] = df['胜率'].str.rstrip('%').astype(float) / 100
df.胜 = df.胜.astype(float)
df.负 = df.负.astype(float)
df.净胜 = df.净胜.astype(float)

# 按“所属联盟”分组，计算统计值
result = df.groupby('所属联盟').agg({
    '胜': ['max', 'min', 'mean', 'std'],
    '负': ['max', 'min', 'mean', 'std'],
    '胜率新': ['max', 'min', 'mean', 'std'],
    '净胜': ['max', 'min', 'mean', 'std']
})


result
```

```python
# 柱状图显示东西赛区的胜率（百分比）均值

# 将胜率转换为浮点数
df['胜率(小数)'] = df['胜率'].str.rstrip('%').astype(float).apply(lambda x: x / 100)

# 根据赛区分组并计算胜率均值
mean_win_rate_by_division = df.groupby('所属联盟')['胜率(小数)'].mean()

# 绘制柱状图
plt.figure(figsize=(8, 6))
mean_win_rate_by_division.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Mean Win Rate by Division')
plt.xlabel('Division')
plt.ylabel('Mean Win Rate (%)')
plt.xticks(rotation=0)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

```python
# 将空值的列'老年人(%)'、'儿童(%)'中数据用前值替换；'男性(%)'、'女性(%)'用均值替换

df['老年(%)'] = df['老年(%)'].fillna(method='ffill')
df['儿童(%)'] = df['儿童(%)'].fillna(method='ffill')

df['男性(%)'] = df['男性(%)'].fillna(df['男性(%)'].mean())
df['女性(%)'] = df['女性(%)'].fillna(df['女性(%)'].mean())

```

```python
# 将除了 时间(年) 列外所有的列的数据改成保留小数点后2位小数
# 提示：采用lambda函数和applymap函数

df.iloc[:, 1:] = df.iloc[:, 1:].applymap(lambda x: f'{x:.2f}')

df
```

```python
# 绘制男女比率差绝对值的曲线图

# 计算男女比率差的绝对值

gender_difference = [abs(male - female) for male, female in zip(df['男性(%)'], df['女性(%)'])]

# 绘制曲线
plt.figure(figsize=(10, 6))

plt.plot(df['时间'], gender_difference, marker='o', label='Absolute Gender Difference (%)', color='b')

# 添加标题和标签
plt.title('Absolute Gender Difference Over Years')
plt.xlabel('Year')
plt.ylabel('Absolute Difference (%)')

# 添加图例
plt.legend()

# 显示图形
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 在一个画布的四个子画布上(两行两列)分别画出：
# 1、近十年老人、儿童、其它（100减去老人、儿童占比）人群比率均值的柱状比较图
# 2、近五年男性、女性比率的柱状比较图
# 3、近十年人口增增长率和出生率的水平柱状图
# 4、2022年老人、儿童、其它的饼图

fig, axes = plt.subplots(2, 2, figsize=(14, 8))

# 1 近十年老人、儿童、其它（100减去老人、儿童占比）人群比率均值的柱状比较图
df_lst10y = df[(df['时间']<=2023) & (df['时间']> (2023 - 10))]
df_lst10y['其他(%)'] = 100 - df_lst10y['老年(%)'] - df_lst10y['儿童(%)']

axes[0, 0].bar(['Elderly', 'Child', 'Other'], [df_lst10y['老年(%)'].mean(),df_lst10y['儿童(%)'].mean(),df_lst10y['其他(%)'].mean()], color='skyblue')
axes[0,0].set_title('Average Population Ratios Over the Past Ten Years')
axes[0,0].set_ylabel('Percentage (%)')

# 2 近五年男性、女性比率的柱状比较图
df_lst5y = df[(df['时间']<=2023) & (df['时间']> (2023 - 5))]
axes[0,1].bar(df_lst5y['时间'],df_lst5y['女性(%)'],label='Female')
axes[0,1].bar(df_lst5y['时间'],df_lst5y['男性(%)'],label='Male',bottom=df_lst5y['女性(%)'])
axes[0,1].set_xlabel('Year')
axes[0,1].set_ylabel('Percentage (%)')
axes[0,1].set_title('Male-Female Ratios Over the Past Five Years')

# 3 近十年人口增增长率和出生率的水平柱状图
axes[1,0].bar(df_lst10y['时间'],df_lst10y['增长率(‰)'],label='Population Growth Rate')
axes10 = axes[1,0].twinx()
axes10.plot(df_lst10y['时间'],df_lst10y['出生率(‰)'],label='Population Birth Rate')
axes[1,0].set_xlabel('Year')
axes[1,0].set_ylabel('Population Growth Rate (%)')
axes10.set_ylabel('Birth Rate (‰)')
axes[1,0].set_title('Population Growth Rate and Birth Rate Over the Past Ten Years')

# 4 2022年老人、儿童、其它的饼图
df_2022 = df[df['时间'] == 2022]

sizes = [df_2022['老年(%)'].values[0],df_2022['儿童(%)'].values[0],100 - df_2022['老年(%)'].values[0] - df_2022['儿童(%)'].values[0]]
labels = ['Elderly', 'Child', 'Other']

axes[1,1].pie(sizes,labels = labels,autopct='%1.1f%%')
axes[1,1].set_title('Distribution of Population in 2022')
axes[1,1].axis('equal')

plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```

```python
# 计算增长率的平均值和标准差
mean_growth_rate = df['增长率(‰)'].mean()
std_growth_rate = df['增长率(‰)'].std()

# 计算异常值的阈值
threshold = mean_growth_rate + 1.5 * std_growth_rate

# 找出增长率异常值的年份
outliers = df[df['增长率(‰)'] > threshold]['时间'].tolist()

print("Outliers in Growth Rate:", outliers)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 构造数据
x = np.linspace(-1 * np.pi, np.pi, 256, endpoint=True)
y1 = np.sin(x)
y2 = np.cos(x)

# 创建画布
plt.figure(figsize=(8, 6))

# 绘制两条折线
plt.plot(x, y1, label='$y=sin(x)$')
plt.plot(x, y2, label='$y=cos(x)$')

plt.xlim(-1.3 * np.pi, 1.3 * np.pi)  # 设置x轴刻度范围
plt.ylim(-1.3, 1.3)  # 设置y轴刻度范围

# 设置x轴刻度
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           ['$-\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$'])

# 设置y轴刻度
plt.yticks([-1, 1])

# 绘制填充区域
plt.fill_between(x, y2, 0, where=(x > -2.5) & (x < -0.5), color='purple', alpha=0.3)  # 紫色区域：(-2.5<x<-0.5)
plt.fill_between(x, 0, y2, where=(x > -2.5) & (x < -0.5), color='purple', alpha=0.3)  # 紫色区域：(-2.5<x<-0.5)
plt.fill_between(x, y1, 0, where=(y1 > 0.5), color='green', alpha=0.3)  # 绿色区域：sinx>0.5

plt.title('绘图实例之SIN()&COS()', color='g')
# 将图例放置在左上角
plt.legend(loc='upper left')

plt.grid(True)  # 添加网格线
plt.axhline(0, color='black', linewidth=0.5)  # 添加水平参考线
plt.axvline(0, color='black', linewidth=0.5)  # 添加垂直参考线

# 设置x轴刻度的位置
plt.gca().spines['bottom'].set_position(('data', 0))

# 设置y轴刻度的位置
plt.gca().spines['left'].set_position(('data', 0))

# 去掉上边框和右边框
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.annotate('cos(-n)=-1', xytext=(-np.pi + 0.7, -1.3), xy=(-np.pi, -1),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.3', ec='black'),
             fontsize=10)

plt.annotate('$sin(2\pi/3) = \sqrt{3}/2$', xytext=(2 * np.pi / 3 + 0.2, np.sqrt(3) / 2 + 0.1),
             xy=(2 * np.pi / 3, np.sqrt(3) / 2),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.0', ec='black'),
             fontsize=10)

plt.text(np.pi / 2 + 0.3, -1.2, 'python_matplotlib', {'size': 12, 'color': 'purple'})

plt.show()
```
