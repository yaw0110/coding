```python
import pandas as pd

# 读取Excel文件中的所有sheet
xlsx = pd.ExcelFile("大数据可视化附件2_1.xlsx")
dfs = []

# 遍历每个sheet
for sheet_name in xlsx.sheet_names:
    # 读取每个sheet的数据
    df = pd.read_excel(xlsx, sheet_name)
    # 添加地区列，并赋值为当前sheet的名称
    df['地区'] = sheet_name
    # 添加到dfs列表中
    dfs.append(df)

# 合并所有DataFrame
merged_df = pd.concat(dfs, ignore_index=True)

# 显示合并后的DataFrame
merged_df
```

```python
# 玫瑰图 timeline 审计重视程度_人均审计支出

def generate_pie_timeline(df):
    attr = df['地区'].unique().tolist()

    tl = Timeline()
    for i in df['年份'].unique().tolist():
        # 筛选出特定年份的数据
        year_data = df[df['年份'] == i]
    
        # 计算每个地区特定年份的平均审计支出
        avg_data = year_data.groupby('地区').agg({'实际利用外资额(万美元)': 'mean'})

        values = avg_data['实际利用外资额(万美元)'].tolist()
    
        pie = (
            Pie()
            .add(
                '实际利用外资额(万美元)',
                [list(z) for z in zip(attr, values)],
                rosetype="radius",
                radius=["30%", "55%"],
            )
            .set_global_opts(title_opts=opts.TitleOpts("实际利用外资额{}年额(万美元)".format(i)))
        )
        tl.add(pie, "{}".format(i))
        tl.load_javascript()
    tl.add_schema(is_auto_play=True, play_interval=2000)
    return tl

pie = generate_pie_timeline(merged_df)
pie.load_javascript()
pie.render_notebook()
pie.render("timeline_audit_exp.html")
```

```python
from pyecharts.charts import Line, Timeline
from pyecharts import options as opts

def generate_stacked_area_timeline(df):
    # 创建一个Timeline对象
    tl_stacked_area = Timeline()
    
    # 遍历数据中的每个年份
    for year in df['年份'].unique().tolist()[::-1]:
        # 筛选出特定年份的数据
        year_data = df[df['年份'] == year]
    
        # 计算每个地区特定年份的审计支出总额和政府财政收入总额
        sum_data = year_data.groupby('地区').agg(
            {
                'GDP(亿元)': 'sum',
                '第二产业增加值': 'sum',
                '第三产业增加值': 'sum',
                '一般公共预算支出（亿元）': 'sum'
            }
        ).reset_index()
    
        # 获取地区和对应的审计支出总额、政府财政收入总额数据
        x_data = sum_data['地区'].tolist()
    
        # 创建堆叠区域图对象
        stacked_area = (
            Line()
            .add_xaxis(x_data)
            .add_yaxis('GDP(亿元)', sum_data['GDP(亿元)'].tolist(), stack="总量", areastyle_opts=opts.AreaStyleOpts(opacity=0.5), label_opts=opts.LabelOpts(is_show=False))
            .add_yaxis('第二产业增加值', sum_data['第二产业增加值'].tolist(), stack="总量", areastyle_opts=opts.AreaStyleOpts(opacity=0.5), label_opts=opts.LabelOpts(is_show=False))
            .add_yaxis('第三产业增加值', sum_data['第三产业增加值'].tolist(), stack="总量", areastyle_opts=opts.AreaStyleOpts(opacity=0.5), label_opts=opts.LabelOpts(is_show=False))
            .add_yaxis('一般公共预算支出（亿元）', sum_data['一般公共预算支出（亿元）'].tolist(), stack="总量", areastyle_opts=opts.AreaStyleOpts(opacity=0.5), label_opts=opts.LabelOpts(is_show=False))
            .set_global_opts(
                title_opts=opts.TitleOpts("总额{}年额".format(year)),
                xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False),  # 设置boundary_gap为False以确保x轴填满
                yaxis_opts=opts.AxisOpts(type_="value"),
                datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")],
                tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross"),
                legend_opts=opts.LegendOpts(pos_left="right", orient="vertical")
            )
        )
    
        # 将堆叠区域图添加到Timeline对象中
        tl_stacked_area.add(stacked_area, "{}".format(year))
    
    # 渲染Timeline对象
    tl_stacked_area.add_schema(is_auto_play=True, play_interval=2000)
    
    return tl_stacked_area
    
generate_stacked_area_timeline(merged_df).render()
```

```python
from pyecharts.charts import Line
from pyecharts import options as opts

def generate_line_timeline(df):
    tl_line = Timeline()
    
    # 遍历数据中的每个年份
    for year in df['年份'].unique().tolist():
        # 筛选出特定年份的数据
        year_data = df[df['年份'] == year]
    
        # 计算每个地区特定年份的平均审计支出
        avg_data = year_data.groupby('地区').agg({'居民消费价格指数((上年=100))': 'mean'}).reset_index()
    
        # 获取地区和对应的平均审计支出数据
        x_data = avg_data['地区'].tolist()
        y_data = avg_data['居民消费价格指数((上年=100))'].tolist()
        
        c = (
            Line()
            .add_xaxis(x_data)
            .add_yaxis(
                "居民消费价格指数((上年=100))", y_data,
                markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_="average")]),
            )
            .set_global_opts(
                title_opts=opts.TitleOpts(title="{}年平均审计支出".format(year)),
                yaxis_opts=opts.AxisOpts(
                    min_=80,   # 设置y轴最小值
                    max_=120   # 设置y轴最大值
                )
            )
        )
    
        # 将柱状图添加到Timeline对象中
        tl_line.add(c, "{}".format(year))
    tl_line.add_schema(is_auto_play=True, play_interval=2000)
    return tl_line
    
# 渲染Timeline对象
generate_line_timeline(merged_df).render()
```

```python
from pyecharts.charts import Funnel, Timeline
from pyecharts import options as opts

def generate_funnel_timeline(df):
# 创建一个Timeline对象
    timeline = Timeline()
    
    # 获取列表
    years = df['年份'].unique().tolist()
    
    # 遍历每个地区
    for year in years:
        # 筛选出特定地区的数据
        year_data = df[df['年份'] == year]
        
        # 创建Funnel对象
        funnel = (
            Funnel()
            .add(
                series_name="创新指数",
                data_pair=[list(z) for z in zip(year_data['地区'], year_data['创新指数'])],
                gap=2,
                tooltip_opts=opts.TooltipOpts(trigger="item", formatter="{a} <br/>{b} : {c}%"),
            )
            # .set_global_opts(title_opts=opts.TitleOpts(title="{}审计支出\n比率漏斗图".format(year)))
        )
        
        # 将Funnel对象添加到Timeline中
        timeline.add(funnel, year)
    timeline.add_schema(is_auto_play=True, play_interval=2000)
    return timeline
# 渲染Timeline对象
generate_funnel_timeline(merged_df).render()
```
