```python
# 玫瑰图 timeline 审计重视程度_人均审计支出

def generate_pie_timeline():
    attr = df['地区'].unique().tolist()

    tl = Timeline()
    for i in df['年份'].unique().tolist():
        # 筛选出特定年份的数据
        year_data = df[df['年份'] == i]
    
        # 计算每个地区特定年份的平均审计支出
        avg_data = year_data.groupby('地区').agg({'审计重视程度_人均审计支出': 'mean'})
    
        values = avg_data['审计重视程度_人均审计支出'].tolist()
    
        pie = (
            Pie()
            .add(
                "审计重视程度_人均审计支出",
                [list(z) for z in zip(attr, values)],
                rosetype="radius",
                radius=["30%", "55%"],
            )
            .set_global_opts(title_opts=opts.TitleOpts("人均审计支出{}年额".format(i)))
        )
        tl.add(pie, "{}".format(i))
        tl.load_javascript()
    tl.add_schema(is_auto_play=True, play_interval=2000)
    return tl
    
generate_pie_timeline().render_notebook()
```

```python
from pyecharts.charts import Line, Timeline
from pyecharts import options as opts

def generate_stacked_area_timeline():
    # 创建一个Timeline对象
    tl_stacked_area = Timeline()
    
    # 遍历数据中的每个年份
    for year in df['年份'].unique().tolist()[::-1]:
        # 筛选出特定年份的数据
        year_data = df[df['年份'] == year]
    
        # 计算每个地区特定年份的审计支出总额和政府财政收入总额
        sum_data = year_data.groupby('地区').agg(
            {
                '审计查出主要问题金额（万元）': 'sum',
                '审计期间整改金额（万元）': 'sum',
                '移送处理涉及金额（万元）': 'sum',
                '审计处理处罚金额（万元）': 'sum'
            }
        ).reset_index()
    
        # 获取地区和对应的审计支出总额、政府财政收入总额数据
        x_data = sum_data['地区'].tolist()
    
        # 创建堆叠区域图对象
        stacked_area = (
            Line()
            .add_xaxis(x_data)
            .add_yaxis("审计查出主要问题金额（万元）", sum_data['审计查出主要问题金额（万元）'].tolist(), stack="总量", areastyle_opts=opts.AreaStyleOpts(opacity=0.5), label_opts=opts.LabelOpts(is_show=False))
            .add_yaxis("审计期间整改金额（万元）", sum_data['审计期间整改金额（万元）'].tolist(), stack="总量", areastyle_opts=opts.AreaStyleOpts(opacity=0.5), label_opts=opts.LabelOpts(is_show=False))
            .add_yaxis("移送处理涉及金额（万元）", sum_data['移送处理涉及金额（万元）'].tolist(), stack="总量", areastyle_opts=opts.AreaStyleOpts(opacity=0.5), label_opts=opts.LabelOpts(is_show=False))
            .add_yaxis("审计处理处罚金额（万元）", sum_data['审计处理处罚金额（万元）'].tolist(), stack="总量", areastyle_opts=opts.AreaStyleOpts(opacity=0.5), label_opts=opts.LabelOpts(is_show=False))
            .set_global_opts(
                title_opts=opts.TitleOpts("审计支出总额与政府财政收入总额{}年额".format(year)),
                xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False),  # 设置boundary_gap为False以确保x轴填满
                yaxis_opts=opts.AxisOpts(type_="value"),
                datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")],
                tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross"),
                legend_opts=opts.LegendOpts(pos_left="right", orient="vertical")
            )
        )
    
        # 将堆叠区域图添加到Timeline对象中
        tl_stacked_area.add(stacked_area, "{}".format(year))
    
    # 渲染Timeline对象
    tl_stacked_area.add_schema(is_auto_play=True, play_interval=2000)
    
    return tl_stacked_area
    
generate_stacked_area_timeline().render_notebook()
```

```python
from pyecharts.charts import Bar, Timeline
from pyecharts import options as opts

def generate_bar_timeline_mul():
    # 创建一个Timeline对象
    tl_bar = Timeline()
    
    # 遍历数据中的每个年份
    for year in df['年份'].unique().tolist():
        # 筛选出特定年份的数据
        year_data = df[df['年份'] == year]
    
        # 计算每个地区特定年份的平均审计支出
        avg_data = year_data.groupby('地区').agg({'审计重视程度_人均审计支出': 'mean'
                                                     , '审计质量指标_整改率': 'mean'
                                                     , '审计效果指标_信息采用率': 'mean'
                                                     , '审计效果指标_建议采纳率': 'mean'
                                                     , '审计质量指标_处罚率': 'mean'}).reset_index()
    
        # 获取地区和对应的平均审计支出数据
        x_data = avg_data['地区'].tolist()
        # y_data = avg_data['审计重视程度_人均审计支出'].tolist()
    
        # 创建柱状图对象
        bar = (
            Bar()
            .add_xaxis(x_data)
            .add_yaxis("审计质量指标_整改率", avg_data['审计质量指标_整改率'].tolist(),
                       label_opts=opts.LabelOpts(is_show=False))  # 隐藏标签
            .add_yaxis("审计效果指标_信息采用率", avg_data['审计效果指标_信息采用率'].tolist(),
                       label_opts=opts.LabelOpts(is_show=False))  # 隐藏标签
            .add_yaxis("审计效果指标_建议采纳率", avg_data['审计效果指标_建议采纳率'].tolist(),
                       label_opts=opts.LabelOpts(is_show=False))  # 隐藏标签
            .add_yaxis("审计质量指标_处罚率", avg_data['审计质量指标_处罚率'].tolist(),
                       label_opts=opts.LabelOpts(is_show=False))  # 隐藏标签
            .set_global_opts(title_opts=opts.TitleOpts("人均审计支出{}年额".format(year)))
    
        )
    
        # 将柱状图添加到Timeline对象中
        tl_bar.add(bar, "{}".format(year))
    
    # 渲染Timeline对象
    tl_bar.add_schema(is_auto_play=True, play_interval=2000)
    return tl_bar

generate_bar_timeline_mul().render_notebook()
```

```python
from pyecharts.charts import Funnel, Timeline
from pyecharts import options as opts

def generate_funnel_timeline():
# 创建一个Timeline对象
    timeline = Timeline()
    
    # 获取列表
    years = df['年份'].unique().tolist()
    
    # 遍历每个地区
    for year in years:
        # 筛选出特定地区的数据
        year_data = df[df['年份'] == year]
        
        # 创建Funnel对象
        funnel = (
            Funnel()
            .add(
                series_name="审计支出比率",
                data_pair=[list(z) for z in zip(year_data['地区'], year_data['审计支出比率'])],
                gap=2,
                tooltip_opts=opts.TooltipOpts(trigger="item", formatter="{a} <br/>{b} : {c}%"),
            )
            # .set_global_opts(title_opts=opts.TitleOpts(title="{}审计支出\n比率漏斗图".format(year)))
        )
        
        # 将Funnel对象添加到Timeline中
        timeline.add(funnel, year)
    timeline.add_schema(is_auto_play=True, play_interval=2000)
    return timeline
# 渲染Timeline对象
generate_funnel_timeline().render_notebook()
```
