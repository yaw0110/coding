```python
import os
import io

# 设置文件夹路径
directory_path = 'data'

# 初始化一个空字符串，用于存储所有文件的内容
all_files_content = io.StringIO()

# 遍历文件夹中的所有文件
for filename in os.listdir(directory_path):
    file_path = os.path.join(directory_path, filename)
    
    # 确保是一个文件而不是文件夹
    if os.path.isfile(file_path):
        with open(file_path, 'r', encoding='utf-8') as f:
            # 读取文件内容并追加到字符串中
            # 可以添加一个分隔符，如换行符，以便区分不同文件的内容
            all_files_content.write(f.read() + '\n')

# 获取拼接后的字符串
combined_content = all_files_content.getvalue()

# 关闭StringIO对象
all_files_content.close()
```

```python
import jieba
import collections
import re
import jieba.posseg as pseg
import pandas as pd


# 定义读取停用词的函数
def read_stop_words(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        stop_words = set(line.strip() for line in file)
    return stop_words

# 停用词表文件路径
stop_words_file = "哈工大停用词表.txt"

# 读取停用词
stop_words = read_stop_words(stop_words_file)

# 使用正则表达式去除文本中的所有特殊字符和数字
clean_text = re.sub(r'[^\w\s]', '', text)  # 去除特殊字符
clean_text = re.sub(r'\d+', '', clean_text)  # 去除数字

# 使用jieba进行分词
words = jieba.lcut(clean_text)

# 去除停用词和单个汉字的词
filtered_words = [word for word in words if word not in stop_words and len(word) > 1]
```

```python
# 分词并词性标注
word_tag_list = list(pseg.cut(clean_text))

# 初始化存放词频的列表
word_freq = collections.Counter(filtered_words)

# 创建DataFrame所需的列表
y1 = list(word_freq.values())  # 词频列表
y2 = list(word_freq.keys())  # 过滤后的词语列表

# 提取过滤后的词语对应的词性
y3 = [flag for word, flag in word_tag_list if word in y2]

# 确保y3的长度与y2相同
if len(y3) != len(y2):
    # 如果长度不同，调整y3的长度，使其与y2相同
    y3 = y3[:len(y2)]

# 打印结果，或者进行其他处理
print("词频: ", y1)
print("词语: ", y2)
print("词性: ", y3)

# 创建DataFrame
z2 = pd.DataFrame({'词频': y1, '分词': y2, '词性': y3})
z2 = z2.sort_values(by='词频', ascending=False)

# 输出到Excel
z2.to_excel('政策文件内容词频统计.xlsx', sheet_name='Sheet1', index=False)
```

```python
from pyecharts.charts import WordCloud
# 创建词云图
wordcloud = WordCloud()

# 添加分词数据
wordcloud.add("", list(zip(y2, y1)), word_size_range=[20, 100])

# 渲染词云图到文件
wordcloud.render("wordcloud.html")
```

```python
from pyecharts import options as opts
from pyecharts.charts import Line
import pandas as pd

# 读取Excel文件中的“时间调整”列
data = pd.read_excel(r'url1.xlsx', 'url')['时间调整']

# 对时间数据进行排序
data = data.sort_values()

# 获取时间的唯一值并排序
unique_times = data.unique()

# 创建一个完整的时间序列，填充缺失的时间为0
time_series = pd.Series(index=unique_times, data=0)

# 将原始数据累加到时间序列中
for time in unique_times:
    time_series.loc[time] = data[data == time].count()

# 创建折线图
line = (
    Line()
    .add_xaxis(time_series.index.tolist())  # 添加x轴数据
    .add_yaxis("政策文件数", time_series.values.tolist())  # 添加y轴数据
    .set_global_opts(
        title_opts=opts.TitleOpts(title="政策文件发布热度"),  # 设置标题
        yaxis_opts=opts.AxisOpts(name="政策文件数量"),  # 设置y轴名称
        xaxis_opts=opts.AxisOpts(name="时间", type_="category"),  # 设置x轴名称和类型
        datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")]  # 设置数据缩放组件
    )
    .set_series_opts(
        # 设置折线图的线条样式和数据标签样式
        itemstyle_opts={
            "normal": {
                "color": "#c23531",  # 设置线条颜色
                "label": {
                    "show": True,  # 设置显示数据标签
                    "position": "right"  # 设置数据标签位置
                },
                "border_color": "rgba(220,220,220,0)",  # 设置线条边框颜色
                "border_width": 1  # 设置线条边框宽度
            },
        }
    )
    .render(r'政策文件发布热度折线图.html')  # 渲染图表到HTML文件
)

# 注意：请将 'url1.xlsx' 替换为您的Excel文件的实际文件名和路径。
```
