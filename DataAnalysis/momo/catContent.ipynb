```python
import requests
import pandas as pd
import re
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('url.csv')
df
```

```python
def get(url,i):
    """
    :param url:政策文件链接 
    :param i: 索引，用来统计
    :return: void
    """
    # 获取静态源代码
    response = requests.get(url)
    response.encoding = 'utf-8' 
    res = response.text
    
    # 筛选正文内容
    pattern = re.compile('id="UCAP-CONTENT"(.*?)用来校验该浏览器是否支持HTML5',re.S)
    soup = BeautifulSoup(pattern.findall(res)[0], 'html.parser')
    
    # 提取所有的<p>标签内容
    paragraphs = soup.find_all('p')
    content_to_save = '\n'.join(p.get_text().strip() for p in paragraphs if p.get_text().strip())
    
    # 命名
    pubtimeStr = df[df['url']==url]['pubtimeStr'].values
    title = df[df['url']==url]['title'].values
    name = (title+"_"+pubtimeStr+".txt").tolist()[0]
    
    safe_filename = re.sub(r'<[^>]+>', '', name)  # 移除HTML标签
    safe_filename = re.sub(r'[\\/:*?"<>|]', '_', safe_filename)  # 移除特殊字符
    
    # 写入文件
    with open('data/'+safe_filename, 'w', encoding='utf-8') as file:
        file.write(content_to_save)
        print(f"{i} 内容已成功保存到 data/{safe_filename} 文件中。")
```

```python
i = 1
for url in df['url'].tolist():
    get(url,i)
    i += 1
```
